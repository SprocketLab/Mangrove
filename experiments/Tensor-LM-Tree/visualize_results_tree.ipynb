```python
import matplotlib.pyplot as plt 
import json 
import os 

def read(fpath):
    s = ""
    with open(fpath,'r') as f:
        s= f.read()
    return s 

out_dir = '../../outputs/tree_runs/0/'
fpaths = os.listdir(out_dir)
fpaths  = [out_dir+fpath for fpath in fpaths]

outs = [json.loads(read(fpath)) for fpath in fpaths ]

import pandas as pd 
df = pd.DataFrame(outs)
#df.head()
```

```python
lst_k = sorted(df['k'].unique())
M = df.groupby('k').mean()
S = df.groupby('k').std()
```

```python
lst_k = sorted(df['k'].unique())
M = df.groupby('k').mean()
S = df.groupby('k').std()

plt.plot(M['acc_mv_dist'],label='Majority Vote',marker='*')
plt.plot(M['acc_pse_dist'],label='Tensor-LM-PSE',marker='s')
#plt.plot(M['acc_1_hot_dist'],label='Tensor-LM-1-Hot',marker='x')
plt.plot(M['acc_fs_dist'],label='Flying-Squid',marker='o')
plt.legend()
plt.xlabel('k')
plt.ylabel('Error (Squared Tree Distance)')
plt.grid()
```

```python
lst_k = sorted(df['k'].unique())
M = df.groupby('k').mean()
S = df.groupby('k').std()

plt.plot(M['acc_mv'],label='Majority Vote',marker='*')
plt.plot(M['acc_pse'],label='Tensor-LM-PSE',marker='s')
plt.plot(M['acc_1_hot'],label='Tensor-LM-1-Hot',marker='x')
plt.legend()
plt.xlabel('k')
plt.ylabel('accuracy(0-1)')
plt.grid()
```

```python
plt.plot(M['acc_pse'],label='Tensor-LM-PSE',marker='s',linestyle='--',color='tab:red',linewidth=2)
plt.plot(M['acc_1_hot'],label='Tensor-LM-1-Hot',marker='d',linestyle='--',color='tab:green',linewidth=2)
plt.plot(M['acc_fs'],label='Flying-Squid',marker='o',linestyle='--',color='tab:orange',linewidth=2)
plt.plot(M['acc_mv'],label='Majority Vote',marker='*',linestyle='--',color='tab:blue',linewidth=2)

plt.legend()
plt.xlabel('Cardinality (k)')
plt.ylabel('Mean Accuracy (0-1)')
plt.ylim(0.55,0.9)
plt.grid()
plt.tight_layout()
plt.savefig('../../plots/figure_mean_acc_tree_all.jpg',dpi=400)

```
